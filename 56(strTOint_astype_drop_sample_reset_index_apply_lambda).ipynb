```python
import pandas as pd
```

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str.replace("[\$\,\.]", "").astype(int)
```

```python
#The line of code below tells to keep the values same, but access the string and replace the "$"dollar and ","commas and "."dots, with ""noting:
import numpy as np
car_sales['Price'] = car_sales['Price'].astype(str).str[:-2].astype(np.int64)
```

```python
car_sales
```

##**Create a column from series:**

```python
#Column from series:
seats_column = pd.Series([5, 5, 5, 5, 5]) #this has 5 entries but will still add because the latter 5 will just add Nan
car_sales["Seats"] = seats_column
```

```python
car_sales #$the missing valies below is because our series  had 5 entires only
```

###**Filling the missing in "Seats" with 5**

> Indented block



```python
car_sales["Seats"].fillna(5, inplace=True) #Filling the missing in "Seats" with 5
```

```python
 car_sales
```

##**Create a column from list:**

```python
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 4.7, 7.6, 8.1, 3.0, 9.1] #this has 10 entries so will still add but if it was 5, there would be an error
car_sales["Fuel per 100KM"] = fuel_economy
car_sales
```

##**Create a column from column:**

```python
car_sales["Total_fuel_used"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100KM"]
```

```python
car_sales["Total_fuel_used_L"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100KM"]
```

```python
car_sales
```

##**Create a column from a single value:**

```python
car_sales["Number of wheels"] = 4
```

```python
car_sales
```

```python
car_sales["Passed road safety"] = True
car_sales
```

##**Remove a column**

```python
car_sales = car_sales.drop("Total_fuel_used", axis=1) #the columns are axis=1 and rows=0  
car_sales
```

##**Shuffle rows in a DataFrame: .sample**

```python
car_sales.sample(frac=0.5) #0.5 is 50% of the data
```

```python
car_sales_shuffled = car_sales.sample(frac=1) #0.5 is 100% of the data
```

```python
car_sales
```

```python
#Only select 20% of data
car_sales_shuffled.sample(frac=0.2)
```

```python
car_sales_shuffled
```

##**Fixing the index again: **

```python
car_sales_shuffled.reset_index() #but this provides an extra column of the index !!
```

```python
car_sales_shuffled.reset_index(drop=True, inplace=True) #as we set inplace=True, that extra column goes away
```

```python
car_sales
```

##**How to apply a function to a column: .apply and lambda ***** **

```python
# Lambda is a keyword in python which is short for an anonymous function
#This means apply this function to x/1.6
#Apply reset x to be = x/1.6
#In this case x is the value in our "Odometer (KM) column"
car_sales["Odometer (KM)"] = car_sales["Odometer (KM)"].apply(lambda x: x / 1.6)
car_sales
```
