#**--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------**

#**OO METHOD vs SUBPLOTS:**

###**Which one should you use? (pyplot vs matplotlib OO method?)**



*   When plotting something quickly, okay to use pylot method
*   When plotting something advanced, okay to use OO method






```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

#**PYPLOT METHOD:**

```python
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             figsize=(10, 6)) #c for color
```

#**--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------**

#**OO METHOD MIXED WITH PYPLOT METHOD:**

```python
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);
#we can access the axis here:
ax.set_xlim([45, 100]);
```

#**--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------**

#**HOW TO MAKE THE OO PLOT A LITTLE BIT NICER**

#**OO METHOD:** use matplotlib documentation

```python
#OO method from scratch:
fix, ax = plt.subplots(figsize=(10, 6))
#Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])
#Customise the plot:
ax.set(title="Heart Disease and cholesterol labels",
       xlabel="Ageeee",
       ylabel="Cholesteroll");
#Add a legend: Adds a chart at the top right about which color indicates what
ax.legend(*scatter.legend_elements(), title="Target"); 

#Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle="--")
```

```python
over_50.head()
```

```python
#Subplot of chol, age, thalach:
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(4, 4))
```

```python
#Subplot of chol, age, thalach:
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10))
#--------------------------------------------------------------------------------------
#Add data to ax0:
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])
#Customise ax0:
ax0.set(title="Heart d and chol",
        xlabel="Agee",
        ylabel="CHOLL")
#Add a legend:
ax0.legend(*scatter.legend_elements(), title="Target")
#Add a meanline:
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");
#--------------------------------------------------------------------------------------

#Add data to ax1:
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])
#Customise ax0:
ax1.set(title="Heart d and chol",
        xlabel="Agee",
        ylabel="THALACHH")
#Add a legend:
ax1.legend(*scatter.legend_elements(), title="Target")
#Add a meanline:
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");
```

```python
#The x-axis is age for both the graphs!! In this case we might wan to share axis!!! 
```

```python
#Subplot of chol, age, thalach:
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True) #SHARE AXIS IS SET TO TRUE ***
#--------------------------------------------------------------------------------------
#Add data to ax0:
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])
#Customise ax0:
ax0.set(title="Heart d and chol",
        # xlabel="Agee", #THIS LINE IS REMOVED AS WE ARE SHARING THE X AXIS OF ax1 ***
        ylabel="CHOLL")
#Add a legend:
ax0.legend(*scatter.legend_elements(), title="Target")
#Add a meanline:
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");
#--------------------------------------------------------------------------------------

#Add data to ax1:
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])
#Customise ax0:
ax1.set(title="Heart d and chol",
        xlabel="Agee",
        ylabel="THALACHH")
#Add a legend:
ax1.legend(*scatter.legend_elements(), title="Target")
#Add a meanline:
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");

#Add a title to the entire figure: **
fig.suptitle("HEART DISEASE ANALYSIS", fontsize=16, fontweight="bold");
```

#**Customising the x and y axis limitations:**

```python
#Subplot of chol, age, thalach:
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True) #SHARE AXIS IS SET TO TRUE ***
#--------------------------------------------------------------------------------------
#Add data to ax0:
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="winter") #changes the solor of dots in scatter
#Customise ax0:
ax0.set(title="Heart d and chol",
        # xlabel="Agee", #THIS LINE IS REMOVED AS WE ARE SHARING THE X AXIS OF ax1 ***
        ylabel="CHOLL")

#Add a legend:
ax0.legend(*scatter.legend_elements(), title="Target")
#Add a meanline:
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");
#--------------------------------------------------------------------------------------

#Add data to ax1:
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="plasma")#changes the solor of dots in scatter
#Customise ax0:
ax1.set(title="Heart d and chol",
        xlabel="Agee",
        ylabel="THALACHH")
#Add a legend:
ax1.legend(*scatter.legend_elements(), title="Target")
#Add a meanline:
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");

#Add a title to the entire figure: **
fig.suptitle("HEART DISEASE ANALYSIS", fontsize=16, fontweight="bold");
```

```python
#Subplot of chol, age, thalach:
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True) #SHARE AXIS IS SET TO TRUE ***
#--------------------------------------------------------------------------------------
#Add data to ax0:
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="winter") #changes the solor of dots in scatter

#Customise ax0:
ax0.set(title="Heart d and chol",
        # xlabel="Agee", #THIS LINE IS REMOVED AS WE ARE SHARING THE X AXIS OF ax1 ***
        ylabel="CHOLL")

#Change the x-axis limit
ax0.set_xlim([50, 80]) #sets x axis limit
ax0.set_ylim([60, 200]) #sets x axis limit

#Add a legend:
ax0.legend(*scatter.legend_elements(), title="Target")

#Add a meanline:
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");
#--------------------------------------------------------------------------------------

#Add data to ax1:
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="plasma")#changes the solor of dots in scatter
#Customise ax0:
ax1.set(title="Heart d and chol",
        xlabel="Agee",
        ylabel="THALACHH")
#Add a legend:
ax1.legend(*scatter.legend_elements(), title="Target")
#Add a meanline:
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");

#Add a title to the entire figure: **
fig.suptitle("HEART DISEASE ANALYSIS", fontsize=16, fontweight="bold");
```
