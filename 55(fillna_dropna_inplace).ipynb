```python
import pandas as pd
car_sales_missing = pd.read_csv("car-sales-missing-data.csv")
car_sales_missing #Nan represents missing values
```

##**What if we wanted to fill a column?** *italicised text*

**Fill with mean:**

```python
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
```

###**Even after filling, the Nan is still there !!!**

```python
car_sales_missing.head() 
```

###**In this case we can do two things: **
car_sales_missing["Odometer"] = car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())

----------------OR------------------------------------------

Use the **inplace** parameter:

```python
#Initially the .fillna(inplace=False), we have to set it to True
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean(), inplace=True)
```

```python
car_sales_missing.head() #Now you can see that the missing values are filled !!
```

##**What if we wanted to get rid of the missing values?**

```python
car_sales_missing.dropna()
```

```python
car_sales_missing #there are still Nan values!
```

```python
#To fix that:
car_sales_missing.dropna(inplace=True) #there are still Nan values!
```

```python
car_sales_missing #Mow there are no missing values / Nan
```
